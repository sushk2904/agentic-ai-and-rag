### Models Integration with OpenAI, Google Gemini and Groq

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("groq:llama-3.3-70b-versatile")
model
```

```python
#invoking the model
response = model.invoke("Hello How are you?")
response.content
```

### Google Gemini Integration

```python
import os
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash-lite")
response = model.invoke("Why do parrots talk")
response.content
```

```python
#ChatOpenAI
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="gpt-4.1")
response = model.invoke("How are you?")
response.content
```

```python
import os
from langchain_google_genai import GoogleGenerativeAI
model =  GoogleGenerativeAI(model = "gemini-2.5-flash-lite")
response =  model.invoke("How are you?")
response
```

```python
from langchain_groq import ChatGroq
model = ChatGroq(model = "qwen/qwen3-32b")
response = model.invoke("How are you? How many parameters have you been trained on?")
response.content
```

### Streaming

```python
model.invoke("Write me a 200 words paragraph on Artificial Intelligence")
```

```python
for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text, end="|", flush=True)
```

```python
for chunk in model.stream("Why do parrots have colour feathers?"):
    print(chunk.text, end= "")
```

```python
response = model.batch([
    "How can I crack 1 CR package as a fresher in 2026 working in IT field?",
    "What is the highest amount of package offered to a fresher who codes in Python and is a fresher in 2026?",
    "I'm not from a good college,  I'm from a tier 3 college and beginner at DSA and starting my 3rd year in 2026, how can I get a package of 1 CR as a fresher?",
], config = {"max_concurrency":5, #This limits the maximum of concurrent requests to 5.
}
               

)
for batch in response:
    print(batch.content, end="\n", flush=True)
```
